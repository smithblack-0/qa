```python
# About how to extract path info

source = r'C:\Users\chris\PycharmProjects\fastai\Data\lambada-dataset'
files = core.get_text_files(source)

texts = []
for i in range(10):
    file = files[i]
    with open(file, errors='ignore') as f:
        text = f.read()
        text = nltk.word_tokenize(text)
    texts.append(nltk.Text(text))

texts = nltk.TextCollection(texts)
texts.vocab()
```
